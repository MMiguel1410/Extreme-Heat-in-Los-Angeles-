```python
# Importing pysal to reclassify data
import pysal as ps
import mapclassify
n_classes = 5
```

```python
# Importing data
df=gpd.read_file('CVA.geojson')
```

```python
# Trimming data
df_trimmed=df[['Census_Tract', 'Population', 'Children', 'Older_Adults', 'Older_Adults_Living_Alone', 'Asthma', 'Cardiovascular_Disease', 'Disability', 'No_Health_Insurance', 'geometry']]
```

```python
# Creating bins to reclassify
classifier=mapclassify.NaturalBreaks.make(k=5)
```

```python
# Adding my new reclassifyed coloumns 
df_trimmed['children_classes']=df_trimmed[["Children"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['elderly_classes']=df_trimmed[["Older_Adults"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['elderly_alone_classes']=df_trimmed[["Older_Adults_Living_Alone"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['asthma_classes']=df_trimmed[["Asthma"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['cardiovascular_classes']=df_trimmed[["Cardiovascular_Disease"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['disabeled_classes']=df_trimmed[["Disability"]].apply(classifier)
df_trimmed
```

```python
df_trimmed['no_health_classes']=df_trimmed[["No_Health_Insurance"]].apply(classifier)
df_trimmed
```

```python
# Calculating my new vulnerability score
df_trimmed['Vuln_Score']=df_trimmed['children_classes']+df_trimmed['elderly_classes']+df_trimmed['elderly_alone_classes']+df_trimmed['asthma_classes']+df_trimmed['cardiovascular_classes']+df_trimmed['disabeled_classes']+df_trimmed['no_health_classes']
df_trimmed
```

```python
# Mapping
m=folium.Map(location=[34.2,-118.2],
             zoom_start=9,
             tiles='CartoDB positron',
             attribute='CartoDB')
```

```python
folium.Choropleth(
geo_data=df_trimmed,
    data=df_trimmed,
    key_on='feature.properties.Census_Tract',
    columns=['Census_Tract', 'Vuln_Score'],
    fill_color='YlOrBr',
    line_weight=0.1,
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name='Health Vulnerability Index').add_to(m)
m
```

```python
m.save('Vuln_Index.html')
```
